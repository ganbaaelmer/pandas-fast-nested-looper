```python

@numba.jit(parallel=True)
def pandas_fast_nested_looper(file2_column_A_numpy, file1_column_A_numpy, file1_column_B_numpy, progress):
    file2_column_B_list = [0] * file2_column_A_numpy.shape[0]
    
    for i in range(num_iterations):
        progress[0].update()

        for j in range(num_iterations2):
            if file2_column_A_numpy[i] == file1_column_A_numpy[j]:
                file2_column_B_list[i] = file1_column_B_numpy[j]
                
            else:
                file2_column_B_list[i] = 99999999

            progress[1].update(1)
            
        print(1)
        # reset the second progress bar to 0
        progress[1].set(0)

    print('loop done!')
    
    return file2_column_B_list


if __name__ == "__main__":
    # with ProgressBar(total=num_iterations) as progress:
    with ProgressBar(total=num_iterations, ncols=80) as numba_progress1, ProgressBar(total=num_iterations2, ncols=80) as numba_progress2:
        file2_column_B_list = pandas_fast_nested_looper(file2_column_A_numpy, file1_column_A_numpy, file1_column_B_numpy, (numba_progress1, numba_progress2))

print('loop done!')
```
